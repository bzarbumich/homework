```python

import pandas as pd
import plotly.express as px

df = pd.read_csv('/content/drive/MyDrive/DATA/olympic_medals.csv')

df = df[df['Year'] >= 2000]

fig = px.parallel_categories(df, dimensions=['Year', 'Gender', 'Medal'])

fig.update_layout(
    title="Proportions of Medal Types for Each Gender Since Year 2000",
    xaxis_title="Year",
    yaxis_title="Proportions"
)

fig.show()

```

```python
!pip install seaborn statsmodels

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load the Cars93 dataset
url = 'https://vincentarelbundock.github.io/Rdatasets/csv/MASS/Cars93.csv'
cars93 = pd.read_csv(url)

def plot_smooth_final(method, color, title):
    sns.lmplot(x='Price', y='Fuel.tank.capacity', data=cars93,
               line_kws={'color': color},
               scatter_kws={'color': 'grey'},
               lowess=(method == 'lowess'),
               robust=(method == 'rlm'),
               logistic=(method == 'logit'),
               truncate=True, ci=95 if method != 'lowess' else None)  # 'ci' parameter does not work with 'lowess'

    plt.xlabel('Price (USD)')
    plt.ylabel('Fuel-tank capacity (US gallons)')
    plt.title(title, fontsize=14, color=color)
    plt.show()

# Plot with different methods and configurations
methods = ['lm', 'rlm', 'lowess']
colors = ['#8fe388', '#fe8d6d', '#7c6bea']
titles = ['Method: lm', 'Method: rlm', 'Method: lowess']

for method, color, title in zip(methods, colors, titles):
    plot_smooth_final(method, color, title)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create DataFrame
df = pd.DataFrame({'High Temp': [28, 30, 34, 38, 45, 42,
                                 38, 35, 32, 28, 25, 21],
                   'Low Temp': [22, 26, 30, 32, 41, 38,
                                32, 31, 28, 22, 15, 15],
                   'Avg Temp': [25, 28, 32, 35, 43, 40,
                                35, 33, 30, 25, 20, 18]},
                  index=['Jan', 'Feb', 'Mar', 'Apr', 'May',
                         'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                         'Nov', 'Dec'])


# create stacked bar chart for monthly temperatures
df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])

# labels for x & y axis
plt.xlabel('Months')
plt.ylabel('Temp ranges in Degree Celsius')

# title of plot
plt.title('Monthly Temperatures in a year')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

elections_df = pd.read_csv('/content/drive/MyDrive/DATA/2016elections.csv')

selected_states = elections_df[elections_df['st'].isin(['AR', 'MI', 'CA', 'WI'])]

selected_states.set_index('state')[['pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other']].plot(kind='bar', stacked=True)
plt.xlabel('State')
plt.ylabel('Vote Percentage')
plt.title('2016 Election Vote Percentages by State')
plt.show()
```
